# **Removing Outlier**

```python
data
```

```python
# dropping NA value 
df1 = data.dropna()
print(df1)
```

```python
df1
```

```python
df2=df1.drop(['type', 'region'], axis=1)
```

```python
df2
```

```python
Q1=df2.quantile(0.25)
Q3=df2.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
new_df2=df2[~((df2<Q1-(1.5*IQR))|(df2>(Q3+(1.5*IQR)))).any(axis=1)]
new_df2
```
